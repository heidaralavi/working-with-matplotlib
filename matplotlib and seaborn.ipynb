```python
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(12, 8), sharex=True, sharey=True)
fig.suptitle('this is 3 column in 2 row subplot and in 1 figure')
ax1 = axes[1,1]
ax1.set_title('sub one')
ax1.set_xlabel('label x1')
ax1.set_ylabel('label y1')
plt.tight_layout()
plt.show()
```

```python
fig, axe = plt.subplots(nrows=1, ncols=3, figsize=(11, 5))
# data x and y axis for seaborn
x= np.random.randn(200)
y = np.random.randn(200)
  
# Kde for x var
sns.kdeplot(x= x , ax=axe[0])
sns.kdeplot(x= y , ax=axe[1])
sns.kdeplot(x=x ,y=y , ax=axe[2])
plt.tight_layout()
plt.show()
```
